```python
import numpy as np
import pandas as pd
import re

# import the input from txt file
list_of_lines = []
with open("input.txt") as fp:
    for line in fp:
        list_of_lines.append(line.strip())
print(list_of_lines)
print(len(list_of_lines))
lol = list_of_lines
```

```python
# input looks like this:
# 'Game 1: 1 green, 4 blue; 1 blue, 2 green, 1 red; 1 red, 1 green, 2 blue; 1 green, 1 red; 1 green; 1 green, 1 blue, 1 red', 'Game 2: 2 blue, 2 red, 6 green; 1 red, 6 green, 7 blue; 10 green, 8 blue, 1 red; 2 green, 18 blue, 2 red; 14 blue, 3 green, 1 red; 8 green, 1 red, 9 blue',
# parse the input into 3D array of numbers
# possible colors are red, green and blue

# regex for searching for the nummber of colors in the line
import re
line = '1 green, 4 blue'
# regex to find red
find_red = re.compile(r'(\d+) red')
find_green = re.compile(r'(\d+) green')
find_blue = re.compile(r'(\d+) blue')
# shape of games array  list[list[list[int]]]
games  = games  = [[]for i in range(100)]
l = [s.split(": ")[1] for s in lol]
for i, game in enumerate(l):
    for j, move in enumerate(game.split("; ")):
        print(move)
        red_match = find_red.search(move)
        n_red = int(red_match.group(1)) if red_match else 0
        green_match = find_green.search(move)
        n_green = int(green_match.group(1)) if green_match else 0
        blue_match = find_blue.search(move)
        n_blue = int(blue_match.group(1)) if blue_match else 0
        games[i].append([n_red, n_green, n_blue])
        print(games)
    print()
```

```python
games_maxes = [[] for i in range(100)]
# go trough each game and replace the list of moves with the maximum number of colors
for i, game in enumerate(games):
    print(game)
    max_red = 0
    max_green = 0
    max_blue = 0
    for move in game:
        if move[0] > max_red:
            max_red = move[0]
        if move[1] > max_green:
            max_green = move[1]
        if move[2] > max_blue:
            max_blue = move[2]
    print([max_red, max_green, max_blue])
    games_maxes[i] = [max_red, max_green, max_blue]

```

```python
# The Elf would first like to know which games would have been 
# possible if the bag contained only 12 red cubes, 13 green cubes, and 14 blue cubes?
sum_of_ids = 0
for i, game in enumerate(games_maxes):
    if (game[0] <= 12) and (game[1] <= 13) and (game[2] <= 14):
        print(i, game)
        sum_of_ids += (i + 1)
sum_of_ids
```

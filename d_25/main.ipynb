```python
# First networkx library is imported  
# along with matplotlib 
import networkx as nx 
import matplotlib.pyplot as plt 
```

```python
# Defining a Class 
class GraphVisualization: 
   
    def __init__(self): 
          
        # visual is a list which stores all  
        # the set of edges that constitutes a 
        # graph 
        self.visual = [] 
          
    # addEdge function inputs the vertices of an 
    # edge and appends it to the visual list 
    def addEdge(self, a, b): 
        temp = [a, b] 
        self.visual.append(temp) 
          
    # In visualize function G is an object of 
    # class Graph given by networkx G.add_edges_from(visual) 
    # creates a graph with a given list 
    # nx.draw_networkx(G) - plots the graph 
    # plt.show() - displays the graph 
    def visualize(self): 
        G = nx.Graph() 
        G.add_edges_from(self.visual) 
        nx.draw_networkx(G)
        plt.figure(1,figsize=(20,20)) 
        plt.show() 
  
# Driver code 
G = GraphVisualization() 

for line in lol:
    # split the line into the node and the connections
    node, connections_s = line.split(": ")
    # add the node to the set of nodes
    nodes.add(node)
    # split the connections into a list
    connections_l = connections_s.split()
    # go through the connections
    for c in connections_l:
        nodes.add(c)
        # add the connection to the set of connections
        G.addEdge(node, c)
G.visualize() 
```

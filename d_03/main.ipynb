```python
import numpy as np
import pandas as pd

# import the input from txt file
list_of_lines = []
with open("input.txt") as fp:
    for line in fp:
        list_of_lines.append(line.strip())
print(list_of_lines)
print(len(list_of_lines))
lol = list_of_lines
```

```python
symbols = ['@', '*',  '%', '-', '&', '+', '/', '#', '$', '=']
```

```python
print(len(lol))
print(len(lol[0]))
```

```python
# strategie:
# make a boolean 2d array. mark all fields as false
# go through the symbols and mark all adjacent fields as true
```

```python
length = len(lol[0])
valid_fields = np.array([[False for x in range(length)] for y in range(length)])
valid_fields.shape
```

```python

def validate_number(y,x):
    # go left and right and validate digits until I get to a symbol or dot
    # validate this number
    print("validate ", y, x, lol[y][x])
    if lol[y][x].isdigit() == False:
        return False
    valid_fields[y][x]=True
    # look left
    x1 = x-1
    while x1>=0 and lol[y][x1].isdigit():
        valid_fields[y][x1]=True
        x1 -= 1
    # look right
    x1 = x+1
    while x1<length and lol[y][x1].isdigit():
        valid_fields[y][x1]=True
        x1 += 1
    return True
        
        
```

```python
# go through all lines and act on all symbols
# on my input there is no symbol on the border of the array, so I don't have to take care about the borders

for i,line in enumerate(lol):
    for j, p in enumerate(line):
        if p in symbols:
            print(i,j, p)
            # look left
            if j>0 and lol[i][j-1] not in symbols:
                validate_number(i,j-1)
            # look top left
            if j>0 and i>0 and lol[i-1][j-1] not in symbols:
                validate_number(i-1,j-1)
            
            # look top
            if i>0 and lol[i-1][j] not in symbols:
                validate_number(i-1,j)
            # look top right
            if i>0 and j<length-1 and lol[i-1][j+1] not in symbols:
                validate_number(i-1,j+1)
            # look right
            if j<length-1 and lol[i][j+1] not in symbols:
                validate_number(i,j+1)
            # look bottom right
            if i < length-1 and j<length-1 and lol[i+1][j+1] not in symbols:
                validate_number(i+1,j+1)
            # look bottom
            if i < length-1 and lol[i+1][j] not in symbols:
                validate_number(i+1,j)
            # look bottom left
            if i < length-1 and j>0 and lol[i+1][j-1] not in symbols:
                validate_number(i+1,j-1)
```

# part two
A gear is any * symbol that is adjacent to exactly two part numbers. Its gear ratio is the result of multiplying those two numbers together.
This time, you need to find the gear ratio of every gear and add them all up so that the engineer can figure out which gear needs to be replaced.

This is tricky, because  
123  
.*.  
...  
has 3 adjasent digits, but only one adjasent number
possibilities for 2 adjasent numbers is 
1.1
.*.
...

...
.*.
1.1

or 
one of ....

Can I just make a function, that younts adjasent numbers?

```python
# validate number now returns the number itself or returns 1
def get_number_or_1(y,x):
    # go left and right and validate digits until I get to a symbol or dot
    if lol[y][x].isdigit() == False:
        return 1
    valid_fields[y][x]=True
    # look left
    x0 = x-1
    while x0>=0 and lol[y][x0].isdigit() and x0>=0:
        valid_fields[y][x0]=True
        x0 -= 1
    # look right
    x1 = x+1
    while x1<length and lol[y][x1].isdigit() and x1<length:
        valid_fields[y][x1]=True
        x1 += 1
    print("found ", lol[y][x0+1:x1], " in line ", y)
    return int(lol[y][x0+1:x1])
```

```python
def count_adjasent_numbers_arround(y,x):
    num = 0
    # start top left
    # count top row
    # count one if all 3 are digits
    if lol[y-1][x-1].isdigit() and lol[y-1][x].isdigit() and lol[y-1][x+1].isdigit():
        num += 1
    else:
        num += lol[y-1][x-1].isdigit()
        num += lol[y-1][x+1].isdigit()
    # count middle row
    num += lol[y][x-1].isdigit()
    num += lol[y][x+1].isdigit()
    # count bottom row
    if lol[y+1][x-1].isdigit() and lol[y+1][x].isdigit() and lol[y+1][x+1].isdigit():
        num += 1
    else:
        num += lol[y+1][x-1].isdigit()
        num += lol[y+1][x+1].isdigit()
    return num
        
```

```python
length = len(lol[0])
valid_fields = np.array([[False for x in range(length)] for y in range(length)])
sum_of_gear_ratios = 0
for i,line in enumerate(lol):
    for j, p in enumerate(line):
        if p == "*":
            num_numbers = count_adjasent_numbers_arround(i,j)
            
            if num_numbers == 2:
                print("found Gear at ", i, j)
                valid_fields[i][j] = True
                pass
                y = i
                x = j
                gear_ratio = 1
                if lol[y-1][x-1].isdigit() and lol[y-1][x].isdigit() and lol[y-1][x+1].isdigit():
                    gear_ratio *= get_number_or_1(y-1,x-1)
                else:
                    gear_ratio *= get_number_or_1(y-1,x-1)
                    gear_ratio *= get_number_or_1(y-1,x+1)
                # count middle row
                gear_ratio *= get_number_or_1(y,x-1)
                gear_ratio *= get_number_or_1(y,x+1)
                # count bottom row
                if lol[y+1][x-1].isdigit() and lol[y+1][x].isdigit() and lol[y+1][x+1].isdigit():
                    gear_ratio *= get_number_or_1(y+1,x-1)
                else:
                    gear_ratio *= get_number_or_1(y+1,x-1)
                    gear_ratio *= get_number_or_1(y+1,x+1)
                sum_of_gear_ratios += gear_ratio
                print("found gear ratio: ", gear_ratio)

print("sum of gear ratios: ", sum_of_gear_ratios)
```

Each line of text corresponds to the position and velocity of a single hailstone. The positions indicate where the hailstones are right now (at time 0). The velocities are constant and indicate exactly how far each hailstone will move in one nanosecond.  

Each line of text uses the format px py pz @ vx vy vz. For instance, the hailstone specified by 20, 19, 15 @ 1, -5, -3 has initial X position 20, Y position 19, Z position 15, X velocity 1, Y velocity -5, and Z velocity -3. After one nanosecond, the hailstone would be at 21, 14, 12.

```python
from itertools import combinations
```

```python
for l0,v1 in combinations(hv, 2):
    # find intersection of two lines at x
    # x = (b1-b0)/ (a0 - a1)
    try:
        x = (v1[1]-l0[1])/(l0[0]-v1[0])
    except ZeroDivisionError:
        print("parallel lines")
        if l0[1] == v1[1]:
            print("#### coincident lines")
    print(f"x = {x}, y = {l0[0]*x + l0[1]}")
```

```python
# But I can only count the intersections in the future not the past
max_p = 400000000000000
min_p = 200000000000000
count = 0
for (v0,l0), (v1, l1) in combinations(zip(hv, hl), 2):
    # find intersection of two lines at x
    # x = (b1-b0)/ (a0 - a1)
    try:
        x = (v1[1]-v0[1])/(v0[0]-v1[0])
    except ZeroDivisionError:
        print("parallel lines")
        if v0[1] == v1[1]:
            print("#### coincident lines")
        continue
    y = v0[0]*x + v0[1]
    
    # now check if not only the lines meet, but also the beam meats
    # this means, that the hails path must cross in the future not in their past
    # check if the intersection is in the future of beam 0
    check = True
    # if intersection is right of start then vx must be positive
    # if intersection is left of start then vx must be negative
    # --> vx should have same sign as x - px0
    # --> vx * (x - px0) > 0
    if l0[3] * (x - l0[0]) < 0:
        check = False
    # check if the intersection is in the future of beam 1
    if l1[3] * (x - l1[0]) < 0:
        check = False
    if check:
        # now check if the intersection is inside the bounding box
            if (min_p < x < max_p) and (min_p < y < max_p):
                print(f"x = {x}, y = {y}")
                count += 1
print(count)
```
